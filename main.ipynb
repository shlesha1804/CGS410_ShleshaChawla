```python
import os
import sys
from pathlib import Path
sys.path.insert(0, '/content')

from loader import loadtreebank
from extractor import extractinterveners
from analyzer import computestats
from visualizer import plot_all


TARGETDEPS = ["nsubj", "obj", "amod", "det", "advmod", "obl", "nmod"]

TREEBANKPATHS = {
    "English":  "treebanks/en_ewt-ud-train.conllu",
    "Hindi":    "treebanks/hi_hdtb-ud-train.conllu",
    "Arabic":   "treebanks/ar_padt-ud-train.conllu",
    "Turkish":  "treebanks/tr_imst-ud-train.conllu",
    "Japanese": "treebanks/ja_gsd-ud-train.conllu",
    "Chinese":  "treebanks/zh_gsd-ud-train.conllu",
    "French":   "treebanks/fr_gsd-ud-train.conllu",
    "Spanish":  "treebanks/es_gsd-ud-train.conllu",
    "German":   "treebanks/de_gsd-ud-train.conllu",
}

OUTPUT_DIR = Path("output")
OUTPUT_DIR.mkdir(exist_ok=True)


def main():
    all_records = []

    for lang, path in TREEBANKPATHS.items():
        if not os.path.exists(path):
            print(f"[SKIP] {lang}: file not found at '{path}'")
            continue

        print(f"\n[Processing] {lang} ← {path}")
        sentences = loadtreebank(path)
        print(f"  Loaded {len(sentences)} sentences")

        records = extractinterveners(sentences, lang, TARGETDEPS)
        print(f"  Found {len(records)} non-adjacent dependency pairs with interveners")

        all_records.extend(records)

    if not all_records:
        print("\n[ERROR] No records found. Make sure treebank files exist in treebanks/")
        print("Run:  python download_treebanks.py  to fetch them automatically.")
        sys.exit(1)

    print(f"\n[Total records] {len(all_records)}")


    stats_df = computestats(all_records)
    out_csv = OUTPUT_DIR / "intervener_stats.csv"
    stats_df.to_csv(out_csv, index=False)
    print(f"[Saved] {out_csv}")


    import pandas as pd
    raw_df = pd.DataFrame(all_records)
    raw_out = OUTPUT_DIR / "raw_records.csv"
    raw_df.to_csv(raw_out, index=False)
    print(f"[Saved] {raw_out}")


    plot_all(raw_df, stats_df, OUTPUT_DIR)
    print(f"\n[Done] All outputs in {OUTPUT_DIR}/")


if __name__ == "__main__":
    main()
```
